Instead of training model on entire dataset at once (Which is computationally expensive!) , here we are training on 'block_size' chunks.
The block_size can be any random size, here we are starting with 8.

```python
block_size = 8 
```

```python
train_data[:block_size + 1]
```

What does the above code does is : In context of 18 the very likely next character is 47. 
Similarly, for 18,47 the next very likely character is 56 and so on and so forth.

```python
x = train_data[:block_size]
y = train_data[1:block_size+1]

for i in range(block_size):
    context = x[:i+1]
    target = y[i]
    print(f'{context} : {target}')
```

The above code demonstrates how to prepare input x and target y sequences for training a character-level language model using PyTorch tensors.

This is the length of the context window, set to 8. It determines how many previous characters the model will use to predict the next character.

This tensor contains a sequence of integers representing the encoded characters from the training data, specifically the first block_size characters.

This tensor contains the next character for each position in x, and it is shifted by one position relative to x. For every character in x, the corresponding value in y is the character that follows it in the text.
  
By using a context window or block_size, we train the model to predict the next character given a sequence of previous characters. This approach is computationally efficient and helps the model learn dependencies in the text without processing the entire dataset at once.

Although, it is a one example of a single batch training, the true power of tensors is the parallel processing which is completely independent of each other and trains the model effectively in parallel making the process faster.
