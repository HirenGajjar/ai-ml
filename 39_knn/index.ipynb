```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('data.csv' , usecols=['Age' ,'Pclass' ,'Fare' ,'Survived'])
```

```python
df
```

```python
df.isna().sum()
```

```python
x = df.drop(columns=['Survived'])
```

```python
y = df.Survived
```

```python
x_train,x_test , y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
knn = KNNImputer()
```

```python
x_train_t = knn.fit_transform(x_train)
```

```python
x_test_t = knn.transform(x_test)
```

```python
pd.DataFrame(x_train_t , columns =x_train.columns)
```

```python
lr = LogisticRegression()
lr.fit(x_train_t , y_train)
y_pred = lr.predict(x_test_t)
accuracy_score(y_test, y_pred)
```

The first result is when the default value of neighbors = 5. we can experiment with different values:

```python
def knn_imputer_accuracy_scores(x_train, x_test, y_train, y_test):
    scores = {}
    for n in range(1, 21):
        knn = KNNImputer(n_neighbors=n)
        x_train_t = knn.fit_transform(x_train)
        x_test_t = knn.transform(x_test)
        lr = LogisticRegression()
        lr.fit(x_train_t, y_train)
        y_pred = lr.predict(x_test_t)
        acc = accuracy_score(y_test, y_pred)
        scores[n] = acc
    return scores

knn_imputer_accuracy_scores(x_train, x_test, y_train, y_test)
```

As we can see there is no notable difference with the value of n_neighbors from 1 to 20. so it is better to chose 2 or 4 as it is less calculation and with no effect on performance what so ever.

Now we know that the n_neighbors  =2 or 4 has a great results , we can play with the weight value.

```python
knn = KNNImputer(n_neighbors=4, weights='distance')
x_train_t = knn.fit_transform(x_train)
x_test_t = knn.transform(x_test)
lr = LogisticRegression()
lr.fit(x_train_t, y_train)
y_pred = lr.predict(x_test_t)
accuracy_score(y_test, y_pred)
```

```python
knn = KNNImputer(n_neighbors=4, weights='uniform')
x_train_t = knn.fit_transform(x_train)
x_test_t = knn.transform(x_test)
lr = LogisticRegression()
lr.fit(x_train_t, y_train)
y_pred = lr.predict(x_test_t)
accuracy_score(y_test, y_pred)
```

```python
si = SimpleImputer(strategy='mean')
x_train_t_si = si.fit_transform(x_train)
x_test_t_si = si.transform(x_test)
lr_si = LogisticRegression()
lr_si.fit(x_train_t_si, y_train)
y_pred_si = lr_si.predict(x_test_t_si)
accuracy_score(y_test, y_pred_si)
```
