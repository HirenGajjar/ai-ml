```python
import pandas as pd 
import numpy as np 
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(columns = ['PassengerId' , 'Name' ,'Ticket','Cabin'],inplace=True)
```

```python
x_train, x_test, y_train , y_test = train_test_split(df.drop(columns =['Survived' ]), df.Survived , test_size = 0.2 , random_state = 42)
```

```python
x_train.head(2)
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
si_age = SimpleImputer()
```

```python
si_embarked = SimpleImputer(strategy='most_frequent')
```

```python
x_train_age = si_age.fit_transform(x_train[['Age']])
x_train_embarked = si_embarked.fit_transform(x_train[['Embarked']])
```

```python
x_train_embarked
```

```python
x_test_age = si_age.fit_transform(x_test[['Age']])
x_test_embarked = si_embarked.fit_transform(x_test[['Embarked']])
```

```python
ohe_sex = OneHotEncoder(sparse_output=False , handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False , handle_unknown='ignore')
```

```python
x_train_sex = ohe_sex.fit_transform(x_train[['Sex']])
x_train_embarked = ohe_embarked.fit_transform(x_train_embarked)
```

```python
x_train_embarked
```

```python
x_test_sex = ohe_sex.fit_transform(x_test[['Sex']])
x_test_embarked = ohe_embarked.fit_transform(x_test_embarked)
```

```python
x_train_rem = x_train.drop(columns = ['Sex','Age','Embarked' ] )
x_test_rem = x_test.drop(columns = ['Sex','Age','Embarked' ] )
```

```python
x_train_transform = np.concatenate((x_train_rem , x_train_age , x_train_sex , x_train_embarked) , axis = 1)
```

```python
x_train_transform.shape
```

```python
x_test_transform = np.concatenate((x_test_rem , x_test_age , x_test_sex , x_test_embarked) , axis = 1)
```

```python
x_test_transform.shape
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train_transform, y_train)
```

```python
y_pred = clf.predict(x_test_transform)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```
