```python
import pickle 
```

```python
import numpy as np
```

```python
ohe_sex = pickle.load(open('ohe_sex.pkl' ,'rb'))
```

```python
ohe_embarked = pickle.load(open('ohe_embarked.pkl' , 'rb'))
```

```python
clf = pickle.load(open('clf.pkl' , 'rb'))
```

```python
new_test_input = np.array([2,'male' , 31.0  , 0,0,10.5, 'S'] ,dtype=object).reshape(1,7)
```

```python
new_test_input
```

```python
new_test_input_sex = ohe_sex.transform(new_test_input[:,1].reshape(1,1))
```

```python
new_test_input_sex
```

```python
new_test_input_embarked = ohe_embarked.transform(new_test_input[:,-1].reshape(1,1))
```

```python
new_test_input_sex
```

```python
new_test_input_embarked
```

```python
new_test_input_age = new_test_input[:,2].reshape(1,1)
```

```python
new_test_input_transformed = np.concatenate((new_test_input[:,[0,3,4,5 ]] ,new_test_input_age ,   new_test_input_sex , new_test_input_embarked) , axis = 1)
```
