```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic_toy.csv')
```

```python
df['Age_e1'] = df.Age.fillna(df.Age.mean() + 3*(df.Age.std()))
```

```python
df['Age_e2']  = df.Age.fillna(df.Age.mean() - 3 * (df.Age.std()))
```

```python
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df['Age'].plot(kind='kde', label='Age', linewidth=2)
df['Age_e1'].plot(kind='kde', label='Age_e1', linewidth=2)
df['Age_e2'].plot(kind='kde', label='Age_e2', linewidth=2)
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distribution of Age, Age_e1, and Age_e2')
plt.legend()
plt.show()
```

```python
df['Fare_e1']  = df.Age.fillna(df.Fare.mean() + 3 * (df.Fare.std()))
df['Fare_e2']  = df.Age.fillna(df.Fare.mean() - 3 * (df.Fare.std()))
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
df['Fare'].plot(kind='kde', label='Fare', linewidth=2)
df['Fare_e1'].plot(kind='kde', label='Fare_e1', linewidth=2)
df['Fare_e2'].plot(kind='kde', label='Fare_e2', linewidth=2)
plt.xlabel('Fare')
plt.ylabel('Density')
plt.title('Distribution of Fare, Fare_e1, and Fare_e2')
plt.legend()
plt.show()
```
