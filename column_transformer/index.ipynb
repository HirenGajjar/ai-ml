As fever column has missing values, we selected the fever column from train data and applied simple imputer, same with test data.

For the cough column - the values can have rank (generally strong > mild), we selected train data and applied ordinal encoder and same with test data. 

```python
oe = OrdinalEncoder()
```

```python
x_train_cough = oe.fit_transform(x_train[['cough']])
```

```python
x_train_cough
```

```python
x_test_cough = oe.fit_transform(x_test[['cough']])
```

```python
x_test_cough
```

For city and gender column , that has nominal categorical data (no possible rank), we selected each column from train and test data, applied One Hot encoder.   

```python
ohe = OneHotEncoder()
```

```python
x_train_city
```

```python
x_train_gender = ohe.fit_transform(x_train[['gender']])
```

```python
ohe = OneHotEncoder(drop='first' , sparse_output=False)
```

```python
x_test_city = ohe.fit_transform(x_test[['city']])
```

To concat all the columns, we need to extract the age column

NOW USING COLUMN TRANSFORMER CLASS

```python
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer(transformers=[

    ('t1',SimpleImputer() , ['fever']),
    ('t2',OrdinalEncoder(categories=[['Mild','Strong']]) ,['cough']),
    ('t3',OneHotEncoder(sparse_output=False ,drop='first') ,['gender','city'])
],remainder='passthrough')
```

```python
x_train_transformer_transformed = ct.fit_transform(x_train)
```

```python
x_train_transformer_transformed.shape
```

```python
x_test_transformer_transformed = ct.fit_transform(x_test)
```
