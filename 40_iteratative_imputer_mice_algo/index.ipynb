```python
import pandas as pd 
import numpy as np
```

```python
df = np.round(pd.read_csv('50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
```

```python
df
```

```python
x = df.drop(columns = ['Profit'])
y = df.Profit
y = pd.DataFrame(y)
type(y)
```

```python
import random

df_nan_random = x.copy()

for col_name in df_nan_random.columns:
    nan_indices = random.sample(range(len(df_nan_random)), 10)
    df_nan_random.loc[nan_indices, col_name] = np.nan

df_nan_random.sample()
```

```python
rd_mean = df_nan_random['R&D Spend'].mean()
adm_mean = df_nan_random['Administration'].mean()
mark_mean = df_nan_random['Marketing Spend'].mean()
```

New data frame with missing values & mean calculated

```python
df_nan_random.head()
```

```python
df0 = pd.DataFrame()

df0['rd'] = df_nan_random['R&D Spend'].fillna(rd_mean)
df0['adm'] = df_nan_random['Administration'].fillna(adm_mean)
df0['mark'] = df_nan_random['Marketing Spend'].fillna(mark_mean)
```

```python
df0.head(10)
```

df0 is the step 0 where the missing values is filled with mean of the column

```python
df_1 = df_nan_random['R&D Spend']
df_1.head(2)
```

```python
df_2 = df0[['adm','mark']]
df_2.head(2)
```

```python
df_3  = pd.DataFrame()
```

```python
df_3 = pd.concat([df_1, df_2], axis=1)
df_3.head(3)
```

```python
from sklearn.linear_model import LinearRegression

# Prepare training data (rows where 'R&D Spend' is not null)
train_mask = df_3['R&D Spend'].notnull()
X_train = df_3.loc[train_mask, ['adm', 'mark']]
y_train = df_3.loc[train_mask, 'R&D Spend']

# Prepare test data (rows where 'R&D Spend' is null)
test_mask = df_3['R&D Spend'].isnull()
X_test = df_3.loc[test_mask, ['adm', 'mark']]

# Fit linear regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict missing values
predicted_rd = lr.predict(X_test)

# Fill missing values in df_3['R&D Spend']
df_3.loc[test_mask, 'R&D Spend'] = predicted_rd

df_3.head(10)
```
