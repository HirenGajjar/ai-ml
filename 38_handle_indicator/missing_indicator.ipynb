```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.impute import MissingIndicator
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('train.csv' ,usecols=['Age','Fare',"Survived"])
```

```python
df
```

```python
x= df.drop(columns=['Survived'])
```

```python
y =df.Survived
```

```python
x_train , x_test ,y_train ,y_test  =train_test_split(x,y,random_state=42, test_size=0.42)
```

```python
si = SimpleImputer()
```

```python
x_train_t = si.fit_transform(x_train)
```

```python
x_test_t = si.fit_transform(x_test)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train_t,y_train)
```

```python
y_pred = lr.predict(x_test_t)
```

```python
accuracy_score(y_test, y_pred)
```

Before using the missing Indicator the accuracy score is around 65

```python
mi = MissingIndicator()
```

```python
mi.fit(x_train)
```

```python
mi.features_
```

```python
x_train_m = mi.transform(x_train)
x_test_m = mi.transform(x_test)
```

```python
x_train['Age_na' ]= x_train_m
x_test['Age_na'] = x_test_m
```
